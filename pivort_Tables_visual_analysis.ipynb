```python
import matplotlib.pyplot as plt

# Columns to analyze
date_wise = ['bodyType', 'fuel', 'seatCapacity']

# Loop through each column and plot pie chart
for col in date_wise:
    counts = df[col].value_counts()

    # Create labels with count and percentage
    labels = [f"{label} - {count}" for label, count in zip(counts.index, counts.values)]

    plt.figure(figsize=(6, 6))
    plt.pie(counts.values, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```
